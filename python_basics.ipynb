**Comparison Operators:** Comparison operators are used to compare two values or variables. These operators return Boolean values (True or False) based on whether the comparison is valid or not.

```python
# Assigning values for x and y
x = 10
y = 5
# Compare if x is equal to y
print(x == y)  # False → 10 is not equal to 5
```

```python
# Compare if x is not equal to y
print(x != y)  # True  → 10 is not equal to 5
```

```python
# Compare if x is greater than y
print(x > y)   # True  → 10 is greater than 5
```

```python
# Compare if x is less than y
print(x < y)   # False → 10 is not less than 5
```

```python
# Compare if x is greater than or equal to y
print(x >= y)  # True  → 10 is greater than or equal to 5
```

```python
# Compare if x is less than or equal to y
print(x <= y)  # False → 10 is not less than or equal to 5
```

**Logical Operators:** Logical operators are used to combine conditional statements. They return Boolean values (True or False) based on the conditions being evaluated. Logical operators are mainly used in if statements, loops, and complex condition checks.

```python
x = 10
y = 5

# Using 'and' operator (True if both conditions are true)
print(x > 5 and y < 10)  # True → Both x > 5 and y < 10 are true
```

```python
# Using 'or' operator (True if at least one condition is true)
print(x > 5 or y > 10)   # True → x > 5 is true, so the overall result is true
```

```python
# Using 'not' operator (Reverses the condition)
print(not x > 5)         # False → x > 5 is true, but 'not' reverses it to false
```

**Membership operators**

 Membership operators in Python are used to test if a value or variable is present in a sequence (e.g., string, list, tuple, set, or dictionary). There are two membership operators: in and not in.


*   in: Returns True if the value is found in the sequence, otherwise False.
*   not in: Returns True if the value is not found in the sequence, otherwise False.





```python
# Examples
list1 = [1, 2, 3, 4, 5]
string1 = "hello"
tuple1 = (6, 7, 8)
set1 = {9, 10, 11}
dict1 = {"a": 1, "b": 2}

# in operator
print(3 in list1)
print("e" in string1)
print(9 in set1)
print("c" in dict1)
print(6 in tuple1)

#not in operator
print(6 not in list1)
print("z" not in string1)
print(12 not in set1)
print("a" not in dict1)
print(9 not in tuple1)

```

## **Conditional statements**
Conditional statements are used in programming to make decisions based on specific conditions. These conditions evaluate to either True or False, and the program will execute the code corresponding to the condition that is True.

```python
# IF Statement: A simple conditional statement that executes a block of code if the condition is true.
x = 60
if x > 5:
    print("x is greater than 5")
    # code to execute if condition is true
```

```python
# IF-ELSE Statement: Executes one block of code if the condition is true, and another if it's false.
x = 10
if x > 5:  #code if true
    print("x is greater than 5")
else:       #code if false
    print("x is not greater than 5")
```

```python
# Nested Conditionals: Conditions inside other conditions. Helps to check multiple conditions in sequence.
x = 3
y = 5

if x > 5:
    if y < 10:
        # code if both conditions are true
        print("x is greater than 5 and y is less than 10.")
    else:
        # code if condition1 is true but condition2 is false
        print("x is greater than 5 but y is not less than 10.")
else:
    # code if condition1 is false
    print("x is not greater than 5.")
```

## **Loops**
Loops are control structures that allow you to repeat a block of code multiple times based on certain conditions. They are useful when you need to perform repetitive tasks or iterate over a sequence of items, like lists, strings, or ranges.

```python
# While Loop: Executes a block of code repeatedly as long as a specified condition is true.
# Printing numbers from 1 to 5 using a while loop
i = 1
while i <= 10:
    print(i)
    i += 1  # Incrementing i by 1
```

```python
# For Loop: A for loop is used to iterate over a sequence (like a list, tuple, or string) and execute a block of code for each item in the sequence
# Printing each character in the string using a for loop
word = "Python"
for letter in word:
    print(letter)
```

```python
# Nested Loop: A nested loop is when one loop is inside another loop. This is useful when you want to perform operations on multi-dimensional data (like lists of lists).
# Printing a 2D pattern using nested loops
for i in range(1, 4):  # Outer loop
    for j in range(1, 4):  # Inner loop
        print(f"Row {i}, Column {j}")
```

#The pass Statement
The pass statement serves as a placeholder for future code, preventing errors from empty code blocks.

It's typically used where code is planned but has yet to be written.

```python
def future_function():
    pass  #write code later

# this will execute without any action or error
future_function()
```

```python
n = 10

# use pass inside if statement
if n > 10:
    pass
print("hello")
```

```python
# Parameters and Arguments: Parameters are the variables defined in the function, and arguments are the values passed when calling the function.
# 2 and 3 are arguments, a and b are parameters
def task(a, b):
    return a - b
print(task(2,3))
```

```python
# Multiple Return Values: Functions can return multiple values using commas or as a tuple.
def min_max(values):
    return min(values), max(values)
print(min_max([1, 2, 3, 4, 5]))
```

```python
# Lambdas:
#lambda is used to create anonymous (nameless) functions in Python.
#These are typically used for short, one-line functions— no need to use def

square = lambda x: x ** 2
print(square(2))
```

```python
add = lambda a,b: a+b
sub = lambda x,y: x-y
print(add(3,4))
print(sub(4,5))
```

```python
# Higher-Order Functions: Functions that accept other functions as arguments or return them.
# Higher-order function: executes a given task with a confirmation
# main task of execution
def execute_task(task_func):

    task_func()

# 1st sub task :A simple task: say hello
def greet():
    print("Hello! Hope you're doing well.")

# 2nd task: say welcome
def welcome():
  print("welcome to all")

# Using the higher-order function
execute_task(greet)
execute_task(welcome)
```
