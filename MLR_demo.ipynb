```python
# import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error,r2_score,mean_squared_error
```

```python
# reading data of .csv format
df = pd.read_csv("/content/demo-data.csv")
```

```python
# selecting features and target (X and Y  variable)
X = df.drop(columns= ["Patient_ID","Severity_Score"]) # X variable(independent variable)
y = df["Severity_Score"] # Y variable(dependent variable)
```

```python
# Train-Test-Split
X_train, X_test, y_train,y_test = train_test_split(X,y,test_size = 0.3,random_state= 42 )
```

```python
# build MLR Model
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
# calculating co efficients(beta1-4) and intercept(beta0)
print("\nIntercept(beta0):", model.intercept_)
print("Coefficeints(betai):")
for feature, coef in zip(X.columns,model.coef_):
  print(f"{feature}:{coef}")
```

severity_score= 5.6+(-0.01)Age+(0.0003)Drug_dose+(0.2)ADR+(-0.04)Hospi_days

```python
#6. PREDICTIONS & METRICS
# -------------------------------------------------------
y_pred_test = model.predict(X_test)

print("r2 Score:", r2_score(y_test, y_pred_test))
print("MSE:", mean_squared_error(y_test, y_pred_test))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred_test)))
```
